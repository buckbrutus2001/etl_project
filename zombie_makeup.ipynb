```python
from splinter import Browser
from bs4 import BeautifulSoup

executable_path = {"executable_path": "chromedriver.exe"}    
browser=Browser("chrome", **executable_path, headless=False)

# Visit visitcostarica.herokuapp.com
#url = "https://partycity.com/"
#url = "https://www.partycity.com/search?q=zombie&prefn1=costumeAccessories&prefv1=Makeup%20Special%20Effects"
url = "https://www.liveabout.com/real-and-horrifying-zombie-makeup-tricks-1105891"
browser.visit(url)

# Scrape page into Soup
html = browser.html
soup = BeautifulSoup(html, "html.parser")
print(soup.prettify())
```
